Matplot lib --for visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,'g--')  #plot give line graph
plt.plot(x,z,'r+')
plt.show()  #for showing

```

```python
x=[1,2,3,4,5]
y=[10,15,12,16,18]
plt.scatter(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scatter plot')
plt.show()
```

```python
#using subplot to make multiple plot in same page
plt.subplot(2,2,1) #row,col,index
plt.plot(x,y,color="r")

plt.subplot(2,2,2)
plt.plot([2,3,6,7],[1,2,3,4])

plt.subplot(2,2,3)
plt.plot([2,3,6,7],[1,2,3,4],'b*')

plt.subplot(2,2,4)
plt.plot([2,3,6,7],[1,2,3,4],'r+')


```

```python
#bar plot
x=['eng','hindi','math','science','social sicence']
y=[12,19,13,30,7]
plt.bar(x,y)
plt.show()
```

```python
#pie chart

x=[40,19,13,30,7]
y=['eng','hindi','math','science','social sicence']
plt.pie(x,labels=y,autopct='%1.1f%%')
plt.show()
```

```python
#total bill vs tip data set
tips=sns.load_dataset('tips')
tips.head()
```

```python
print(type(tips))
```

```python
sns.set_theme()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.show()
#plot shows  most of people have given tip in range [2-4] and total price in [10-25]
```

```python
#house price dataset = present in sklearn
from sklearn.datasets import fetch_california_housing
house_dataset=fetch_california_housing()
print(dir(house_dataset))
print(type(house_dataset))
```

```python

# it is not in pandas dataframe from  ..so converting it to pandas

house=pd.DataFrame(house_dataset.data,columns=house_dataset.feature_names)
house['price']=house_dataset.target  # since data do not contain target(ie price) we need to add it manually
house.head()
```

distribution plot

```python
sns.displot(x='price',data=house,kde=True)
```

Correlation matrix

-positive correlation-one increases second also increases

-negative correlation -one increases other decreases

```python
correlation =house.corr();
print(correlation)
```

```python
#construction heat map
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
