 1. What Are Model Parameters?
Model parameters are the internal variables the model learns from the training data.

 Key Facts:
They change during training.

They are learned by minimizing the loss function.

They define the function the model has learned.

 Examples:
Model	Parameters
Linear Regression	Weights

w and bias

b
Neural Network	Weights and biases for every layer
Decision Tree	Split thresholds, decision rules
SVM	Support vectors, coefficients

 2. What Are Hyperparameters?
Hyperparameters are the external settings you define before training the model. They guide how the model trains.

 Key Facts:
Not learned from the data.

Manually set (or tuned) before training.

Control the learning process or model complexity.

 Examples:
Hyperparameter	What it Controls
learning_rate	How big a step to take during gradient descent
batch_size	Number of samples processed before updating weights
epochs	How many times to iterate over the full training data
max_depth (trees)	Depth of decision tree
n_estimators (ensemble)	Number of trees in Random Forest or Gradient Boosting
C, gamma (SVM)	Regularization and kernel influence
hidden_layers, neurons	Architecture of neural networks
